Ejercicio 15

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

Creo una función recursiva que calcula el factorial de n. Si n es 0 o 1 retorna 1, de lo contrario multiplica n por el factorial de n-1.
